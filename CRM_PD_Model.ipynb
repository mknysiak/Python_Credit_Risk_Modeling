```python
# Create a new fit function to calculate p-values ​​for LogisticRegression

from sklearn import linear_model
import scipy.stats as stat

class LogisticRegression_with_p_values:
# this part copies what the LogRef function has
    def __init__(self,*args, **kwargs):
        self.model = linear_model.LogisticRegression(*args, **kwargs)
    
# override original fit function
    def fit(self, X, y):
        self.model.fit(X,y)
        denom = (2.0 * (1.0 + np.cosh(self.model.decision_function(X))))
        denom = np.tile(denom, (X.shape[1],1)).T
        # adding Ridge Penalty to prevent Singular Matrix error
        F_ij = np.dot((X/denom).T,X) + 1e-6
        Cramer_Rao = np.linalg.inv(F_ij)
        sigma_estimates = np.sqrt(np.diagonal(Cramer_Rao))
        z_scores = self.model.coef_[0] / sigma_estimates
        p_values = [stat.norm.sf(abs(x)) * 2 for x in z_scores]
        self.coef_ = self.model.coef_
        self.intercept_ = self.model.intercept_
        self.p_values = p_values
        return
```

```python
# Calculating coefficients, intercept and p_values with new LogisticRegression model

#C: 1.0 – Regularization strength parameter. Smaller values ​​indicate stronger regularization (preventing overfitting).
# fit_intercept: True – indicates that the model has calculated the intercept (the previously mentioned intercept, e.g., your -3.91).
# solver: 'lbfgs' – the default optimization algorithm used to find the best weights for the variables.
# max_iter: 100 – the maximum number of steps (iterations) the algorithm can take to find the optimal solution.

reg = LogisticRegression_with_p_values(max_iter=1000, class_weight=None, random_state=42)
reg.fit(inputs_train, loan_data_targets_train.values.ravel())
# reg.model
```

```python
# Adding p_value to the table with coefficients and intercept

summary_table = pd.DataFrame(columns = ['Feature name'], data = inputs_train.columns.values)
# reg.coef_ are the coefficients for the independent variables
summary_table['Coefficients'] = np.transpose(reg.coef_)
summary_table.index = summary_table.index + 1
# ref.intercept_ is an intercept
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table['p_values'] = np.append(np.nan, np.array(reg.p_values).round(3))

# if p_value below 0.05 - not relevant
# coefficients great to be above 0.2
# summary_table.to_csv('coef_p_values.csv')
```

```python
# Data for GINI and Kolmogorov-Smirnov

df_actual_predicted_probs = df_actual_predicted_probs.sort_values(by='y_hat_test_proba').reset_index(drop=True)
df_actual_predicted_probs['Cumulative N Population'] = df_actual_predicted_probs.index + 1
df_actual_predicted_probs['Cumulative N Good'] = df_actual_predicted_probs['loan_data_targets_test'].cumsum()
df_actual_predicted_probs['Cumulative N Bad'] = df_actual_predicted_probs['Cumulative N Population'] - df_actual_predicted_probs['Cumulative N Good']
df_actual_predicted_probs['Cumulative Perc Population'] = df_actual_predicted_probs['Cumulative N Population']/max(df_actual_predicted_probs.index+1)
df_actual_predicted_probs['Cumulative Perc Good'] = df_actual_predicted_probs['Cumulative N Good']/df_actual_predicted_probs['loan_data_targets_test'].sum()
df_actual_predicted_probs['Cumulative Perc Bad'] = df_actual_predicted_probs['Cumulative N Bad']/(max(df_actual_predicted_probs.index) + 1 - df_actual_predicted_probs['loan_data_targets_test'].sum())
```

```python
# Applying PD model: Scorecard 

df_ref_categories = pd.DataFrame(ref_categories, columns = ['Feature name'])
df_ref_categories['Coefficients'] = 0
df_ref_categories['p_values'] = np.nan

df_scorecard = pd.concat([summary_table, df_ref_categories])
df_scorecard = df_scorecard.reset_index()
df_scorecard['Original feature name'] = df_scorecard['Feature name'].str.split(':').str[0]
# df_scorecard

min_score = 300
max_score = 850

min_sum_coef  = df_scorecard.groupby('Original feature name')['Coefficients'].min().sum()
max_sum_coef  = df_scorecard.groupby('Original feature name')['Coefficients'].max().sum()

df_scorecard['Score - Calculation'] = df_scorecard['Coefficients'] * (max_score - min_score) / (max_sum_coef - min_sum_coef)
df_scorecard.loc[0, 'Score - Calculation'] = (df_scorecard.loc[0, 'Coefficients'] - min_sum_coef) * (max_score - min_score) / (max_sum_coef - min_sum_coef) + min_score
df_scorecard['Score - Preliminary'] = round(df_scorecard['Score - Calculation'],0)

# I check min and max, if too large, I subtract from the variable that I rounded the most, etc.
min_sum_score_prel = df_scorecard.groupby('Original feature name')['Score - Preliminary'].min().sum()
max_sum_score_prel = df_scorecard.groupby('Original feature name')['Score - Preliminary'].max().sum()
# print(max_sum_score_prel)
# print(min_sum_score_prel)

df_scorecard['Score - Final'] = df_scorecard['Score - Preliminary']
df_scorecard['Score - Diff'] = df_scorecard['Score - Final'] - df_scorecard['Score - Calculation']

# idxmin = df_scorecard.groupby('Original feature name')['Score - Final'].idxmax()
# df_scorecard.loc[idxmin]

min_sum_score_final = df_scorecard.groupby('Original feature name')['Score - Final'].min().sum()
max_sum_score_final = df_scorecard.groupby('Original feature name')['Score - Final'].max().sum()
print(max_sum_score_final)
print(min_sum_score_final)

# df_scorecard.to_csv('df_scorecard.csv') 
```
